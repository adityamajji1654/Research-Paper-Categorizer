```python
import numpy as np
from sklearn.metrics import f1_score

def find_best_threshold(y_true, y_pred_probs):
    """
    Find the best threshold for classifying data points into categories to maximize the F1 macro score.

    :param y_true: A binary matrix (numpy array) of shape (n_samples, n_classes) with true labels.
    :param y_pred_probs: A matrix (numpy array) of shape (n_samples, n_classes) with predicted probabilities.
    :return: The best threshold and the corresponding F1 macro score.
    """
    best_threshold = 0.0
    best_f1_score = 0.0
    
    for threshold in np.arange(0.01, 1.0, 0.01):
        # Convert probabilities to binary predictions based on the current threshold
        y_pred = (y_pred_probs >= threshold).astype(int)
        
        # Calculate the F1 macro score
        f1 = f1_score(y_true, y_pred, average='macro')
        
        # Update the best threshold found so far
        if f1 > best_f1_score:
            best_f1_score = f1
            best_threshold = threshold
            
    return best_threshold, best_f1_score
```
