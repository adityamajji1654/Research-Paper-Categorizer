```python
df = pd.read_csv("/kaggle/input/training-data/train.csv", header=0, index_col=0)

df['Title'] = df['Title'].apply(preprocess_text, args=(stop_words, lemmatizer))
df['Abstract'] = df['Abstract'].apply(preprocess_text, args=(stop_words, lemmatizer))
# combine title and abstract into one column and drop original columns
df['Text'] = df['Title'] + ' ' + df['Abstract']

df['Categories'] = df['Categories'].apply(literal_eval)

for col in target_cols:
    df[col] = 0

# Update the DataFrame
for index, row in df.iterrows():
    for category in row['Categories']:
        if category in target_cols:
            df.at[index, category] = 1

main = df.iloc[:, 3:]
main
```

```python
num_categories = len(main.columns) - 1
print(num_categories)
```

```python
texts = main['Text'].tolist()
labels = main[main.columns[1:]].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
model_name = "xlnet-base-cased"
tokenizer = XLNetTokenizer.from_pretrained(model_name)
config = XLNetConfig.from_pretrained(model_name)
base_model = XLNetModel.from_pretrained(model_name)
```

```python
def preprocess_chunk(chunk_texts, chunk_labels):
    encodings = tokenizer(chunk_texts, truncation=True, padding=True,max_length = 128, return_tensors='tf')
    return encodings, chunk_labels
```

```python
model = TFXLNetForSequenceClassification.from_pretrained(model_name, num_labels=num_categories)
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=2e-5)
loss = BinaryCrossentropy()
metrics = [BinaryAccuracy()]
model.compile(optimizer=optimizer, loss=loss, metrics=metrics, run_eagerly=True)
```

```python
chunk_size = 16  # Adjust as needed
for epoch in range(3):  # Number of epochs
    print(f"Epoch {epoch + 1}/{20}:")
    for i in range(0, len(x_train), chunk_size):
        chunk_texts = x_train[i:i + chunk_size]
        chunk_labels = y_train[i:i + chunk_size]
        x_train_encodings, y_train_chunk = preprocess_chunk(chunk_texts, chunk_labels)
        history = model.train_on_batch(dict(x_train_encodings), y_train_chunk)
        if(i%100 == 0):
            print(f"  Chunk {i // chunk_size + 1}/{len(x_train) // chunk_size + 1}: Loss: {history[0]}, Accuracy: {history[1]}")
```
